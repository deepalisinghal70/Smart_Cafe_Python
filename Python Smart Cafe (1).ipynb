```python
# PYTHON SMART CAFE PROJECT

# 1. STORING DATA
beverage_prices = {
    'Matcha Tea': 150,
    'Iced Latte': 200,
    'Mint Frost': 100,
    'Vanilla Swirl': 250,
    'Coffee Brew': 300}

beverage_stock = {
    'Matcha Tea': 10,
    'Iced Latte': 30,
    'Mint Frost': 50,
    'Vanilla Swirl': 20,
    'Coffee Brew': 5}

order = {}
total_bill = 0

# 2. CUSTOMER DETAILS
print("---WELCOME TO THE SMART CAFE---")
customer_name = input("Enter your name: ")
print(f"Hello, {customer_name}! What can I serve you today?\n")

# Display Menu
print("--- MENU ---")
for item in beverage_prices:
    print(f"Beverage: {item} | Price: {beverage_prices[item]} | Stock Left: {beverage_stock[item]}")
print("-----------\n")

# PHASE 1: ITEM SELECTION

print("Kindly begin ordering..")

while True: 
    order_name = input("Enter your order: ")
    if order_name in beverage_stock:
        if beverage_stock[order_name] > 0:
            quantity = int(input("Enter quantity: "))
            if quantity <= beverage_stock[order_name]:
                item_price = beverage_prices[order_name] * quantity
                total_bill += item_price
                beverage_stock[order_name] -= quantity
                order[order_name] = quantity
                print("Order placed successfully! :)")
                break  
            else:
                print("Sorry! Not enough stock available.")
        else:
            print("This beverage is out of stock. Please pick something else.")
    else:
        print("Invalid product name. Please type the name exactly as shown in the menu.")


# =========================================================================
# PHASE 2: THREE-OPTION LOOP (Add more, Remove, or Checkout)
# =========================================================================
while True:
    print("\n--- WHAT WOULD YOU LIKE TO DO NEXT? ---")
    print("A - Add More Items")
    print("B - Remove Items")
    print("C - Checkout")
    
    choice = input("\nEnter your choice: ").upper()

    # OPTION A: ADD MORE ITEMS
    if choice == 'A':
        order_name = input("Enter your order: ")
        if order_name in beverage_stock:
            if beverage_stock[order_name] > 0:
                quantity = int(input("Enter quantity: "))
                if quantity <= beverage_stock[order_name]:
                    item_price = beverage_prices[order_name] * quantity
                    total_bill += item_price
                    beverage_stock[order_name] -= quantity
                    
                    if order_name in order:
                        order[order_name] += quantity
                    else:
                        order[order_name] = quantity
                    print("Item added successfully! :)")
                else:
                    print("Sorry! Not enough stock available. :(")
            else:
                print("This beverage is out of stock. :(")
        else:
            print("Invalid product name.")

    # OPTION B: REMOVE ITEMS
    elif choice == 'B':
        if len(order) == 0:
            print("Your cart is empty!")
            continue
            
        remove_item = input("Enter the item name you want to remove: ")
        if remove_item in order:
            remove_quantity = int(input("Enter the quantity you want to remove: "))
            if remove_quantity <= order[remove_item]:
                order[remove_item] -= remove_quantity
                beverage_stock[remove_item] += remove_quantity
                total_bill -= beverage_prices[remove_item] * remove_quantity
                print("Item removed successfully!")
                
                if order[remove_item] == 0:
                    del order[remove_item]
            else:
                print("You entered a quantity higher than what you ordered.")
        else:
            print("Item not found in your order.")

    # OPTION C: CHECKOUT
    elif choice == 'C':
        print("\nProceeding to Checkout...")
        break  # Breaks this loop to move directly to coupons and final receipt
        
    else:
        print("Invalid choice. Please choose A, B, or C.")


# =========================================================================
# PHASE 3: DISCOUNTS & BILLING
# =========================================================================
print("\nAVAILABLE COUPON:")
print("SAVE10 - 10% OFF")
coupon = input("Enter coupon code or press enter to skip: ")

if coupon == "SAVE10":
    discount = total_bill * 10 / 100
    total_bill -= discount
    print("Coupon applied successfully! :)")
else:
    print("No valid coupon applied.")

# FINAL RECEIPT
print("\n--- FINAL RECEIPT ---")
print("Customer Name:", customer_name)
print("\nItems Purchased:")
for item in order:
    print(f"- {item} (Quantity: {order[item]})")
print("\nTotal Bill:", total_bill)
print("\nThank you for ordering! :)")
```
